#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
merged_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)
merged_df.tail()
```

```python
# Check if any values are null.
nullstatus=merged_df.isnull().sum()
print(nullstatus)
```

```python
# Convert the "invoice_date" to a datetime datatype
merged_df['invoice_date'] = pd.to_datetime(merged_df['invoice_date'])
merged_df.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
merged_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

Total_Products_Sold = merged_df.groupby(['region', 'state', 'city']).agg({'units_sold':'sum'}).reset_index()
Total_Products_Sold.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
Sorted_Total_Products_Sold = Total_Products_Sold.sort_values(by="Total_Products_Sold", ascending=False).reset_index(drop=True)
Sorted_Total_Products_Sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

Total_Products_Sold_PT = merged_df.pivot_table(values  = 'units_sold',
                                               index= ['region', 'state', 'city'],
                                               aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Total_Products_Sold"
Total_Products_Sold_PT.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
Sorted_Total_Products_Sold_PT = Total_Products_Sold_PT.sort_values(by="Total_Products_Sold", ascending=False).reset_index(drop=True)
Sorted_Total_Products_Sold_PT.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
Total_Sales = merged_df.groupby(['region', 'state', 'city',]).agg({'total_sales':'sum'}).reset_index()
Total_Sales.rename(columns={'total_sales': 'Total_Sales'}, inplace = True)
# Show the top 5 results.
Sorted_Total_Sales = Total_Sales.sort_values(by="Total_Sales", ascending=False).reset_index(drop=True)
Sorted_Total_Sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
Total_Sales_PT = merged_df.pivot_table(values  = 'total_sales',
                                       index= ['region', 'state', 'city'],
                                       aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
Total_Sales_PT.rename(columns={'total_sales': 'Total_Sales'}, inplace = True)

# Show the top 5 results.
Sorted_Total_Sales_PT = Total_Sales_PT.sort_values(by="Total_Sales", ascending=False).reset_index(drop=True)
Sorted_Total_Sales_PT.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
Retailer = merged_df.groupby(['retailer', 'region', 'state', 'city',]).agg({'total_sales':'sum'}).reset_index()
Retailer.rename(columns={'total_sales': 'Total_Sales'}, inplace = True)

# Show the top 5 results.
Sorted_Retailer = Retailer.sort_values(by="Total_Sales", ascending=False).reset_index(drop=True)
Sorted_Retailer.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
Retailer_PT = merged_df.pivot_table(values='total_sales',
                                    index=['retailer', 'region', 'state', 'city'],
                                    aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
Retailer_PT.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
Sorted_Retailer_PT = Retailer_PT.sort_values(by='Total_Sales', ascending=False).reset_index(drop=True)
Sorted_Retailer_PT.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_sales = merged_df[merged_df['product']== "Women's Athletic Footwear"]
womens_athletic_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_by_retailer = womens_athletic_footwear_sales.groupby(['retailer', 'region', 'state', 'city',]).agg({'units_sold':'sum'}).reset_index()
womens_athletic_by_retailer.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace = True)

# Show the top 5 results.
sorted_womens_athletic_by_retailer = womens_athletic_by_retailer.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).reset_index(drop=True)
sorted_womens_athletic_by_retailer.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athletic_by_retailer_PT = womens_athletic_footwear_sales.pivot_table(values='units_sold',
                                                                            index=['retailer', 'region', 'state', 'city'],
                                                                            aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_by_retailer_PT.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace = True)

# Show the top 5 results.
sorted_womens_athletic_by_retailer_PT = womens_athletic_by_retailer_PT.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).reset_index(drop=True)
sorted_womens_athletic_by_retailer_PT.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
day_womens_athletic_footwear_sales_pt = womens_athletic_footwear_sales.pivot_table(values= 'total_sales',
                                                                                   index=['invoice_date'],
                                                                                   aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"

day_womens_athletic_footwear_sales_pt.rename(columns={'total_sales':'Total Sales'}, inplace=True)
# Show the table.
day_womens_athletic_footwear_sales_pt
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
day_womens_athletic_footwear_sales_pt=day_womens_athletic_footwear_sales_pt.set_index('invoice_date')
resampled_day_womens_athletic_footwear = day_womens_athletic_footwear_sales_pt.resample('D').sum()
resampled_day_womens_athletic_footwear.head()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_resampled_day_womens_athletic_footwear = resampled_day_womens_athletic_footwear.sort_values(by='Total Sales', ascending=False)
sorted_resampled_day_womens_athletic_footwear.head()
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resampled_weekly_womens_athletic_footwear = day_womens_athletic_footwear_sales_pt.resample('W').sum()
resampled_weekly_womens_athletic_footwear.head()
# Sort the resampled pivot table in descending order on "Total Sales".
sorted_resampled_weekly_womens_athletic_footwear = resampled_weekly_womens_athletic_footwear.sort_values(by='Total Sales', ascending=False)
sorted_resampled_weekly_womens_athletic_footwear.head()
```
